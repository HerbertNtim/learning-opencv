### Creating an image using arrays

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create an array of range(0, 255) with 16 x 16
array = np.arange(0, 256, 1)
print(array)
```

```python
array = array.reshape(16, 16)
print(array)
```

```python
array_8bits = array.astype(np.uint8)
print(array_8bits)
```

```python
plt.axis('off')
plt.imshow(array_8bits, cmap="gray");
```

### Creating an image with random arrays

```python
array2 = np.random.randint(low=0, high=256, size=(16, 16), dtype=np.uint8)
print(array2)
```

```python
plt.axis('off')
plt.imshow(array2, cmap='gray');
```
