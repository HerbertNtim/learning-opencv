### Reading an Image

```python
import numpy as np
import matplotlib.image as matimage
from PIL import Image
import cv2
import matplotlib.pyplot as plt
```

```python
## OpenCV reads image in BGR and returns numpy array
img_cv = cv2.imread('../data/flying_bird.jpg')

## Matplotlib reads image in RGB and returns numpy array
img_mat = matimage.imread('../data/flying_bird.jpg')

## PIL reads image in RGB and returns Object
img_pil = Image.open('../data/flying_bird.jpg')
```

```python
# Open CV
print(img_cv)
```

```python
# Matplotlib 
print(img_mat)
```

```python
img_pil
```

```python
img_pil_array = np.array(img_pil)
```

```python
img_pil_array
```

#### Displaying an Image

```python
# Display image using OpenCV
cv2.imshow("OpenCV BGR Image", img_cv)
cv2.imshow("Matplotlib RGB Image", img_mat)
cv2.imshow("PIL RGB Image", img_pil_array)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Display image using Matplotlib
plt.figure(figsize=(10, 6))
plt.subplot(1, 3, 1)
plt.imshow(img_cv)
plt.axis('off')
plt.title('OpenCV BGR')

plt.figure(figsize=(10, 6))
plt.subplot(1, 3, 2)
plt.imshow(img_mat)
plt.axis('off')
plt.title('Matplotlib RGB')

plt.figure(figsize=(10, 6))
plt.subplot(1, 3, 3)
plt.imshow(img_pil_array)
plt.axis('off')
plt.title('PIL RGB')
```

## COLOR SPACE
. Color Conversion
. Gray Scale

```python
img_cv.shape
```

```python
# Numpy slicing
# BGR OpenCV
b = img_cv[:, :, 0]
g = img_cv[:, :, 1]
r = img_cv[:, :, 2]

```

```python
cv2.imshow("blue channel", b)
cv2.imshow("green channel", g)
cv2.imshow("red channel", r)

cv2.imshow("Original", img_cv)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Opencv split channels
b, g, r = cv2.split(img_cv)
```

```python
cv2.imshow("blue channel", b)
cv2.imshow("green channel", g)
cv2.imshow("red channel", r)

cv2.imshow("Original", img_cv)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
b.shape, g.shape, r.shape, img_cv.shape
```

```python
# Converting color
# bgr to rgb
img_cv_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_cv_rgb)
plt.axis('off')
plt.show()  
```

## Gray Scale

```python
img_gray = cv2.cvtColor(img_cv, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow('Gray Scale Image', img_gray)
cv2.waitKey()
cv2.destroyWindows()
```
