```python
import numpy as np
import cv2
```

```python
# Load the image
img = cv2.imread('../data/flying_bird.jpg')

# Checking the shape and Displaying the image.
print(img.shape)
cv2.imshow("Original Image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Task
1. Shrink the Image
2. Enlarge the image

```python
# Shrink the image
# width=columns, height = rows
img_shrink = cv2.resize(img, dsize=(240, 300), interpolation=cv2.INTER_AREA)

# Enlarge the Image
img_zoom = cv2.resize(img, dsize=(1080, 960), interpolation=cv2.INTER_CUBIC)
```

```python
img_shrink.shape, img_zoom.shape
```

```python
cv2.imshow("Original Image", img)
cv2.imshow("Shrink Image", img_shrink)
cv2.imshow("Enlarge Image", img_zoom)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
